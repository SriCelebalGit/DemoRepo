```sql
select * from orders limit 10
```

```sql
select customer_id,
books,
FILTER(books,i->i.quantity > 2) as multiple_orders
from orders 
```

```sql
select * from (select customer_id,
books,
FILTER(books,i->i.quantity>2) as multiple_orders from
orders) a
where size(a.multiple_orders)>0
```

```sql
select customer_id,
explode(TRANSFORM(
    books,
    b->cast(b.subtotal * 0.8 as INT)
)) AS subtotal_after_discount
from orders
```

```sql
create or replace FUNCTION get_url(email string)
RETURNS string
RETURN concat("https://www.",split(email,"@")[1])
```

```sql
select customer_id, email,
get_url(email) as url
from customers
```

```sql
DESCRIBE FUNCTION EXTENDED get_url
```
