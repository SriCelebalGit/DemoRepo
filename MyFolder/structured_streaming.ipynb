```python
(
spark.readStream
    .table("books")
    .writeStream
    .trigger(availableNow=True)
    .outputMode("append")
    .option("checkpointLocation","dbfs:/Volumes/dbx_catalog/dbx_schema/bookstore/checkpoints/books")
    .table("books_streaming")    
    .awaitTermination()
)
```

```python
%sql
select author,count(1) from books_streaming
group by author
```
