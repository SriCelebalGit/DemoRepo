```sql
create table customers
select * from json.`/Volumes/dbx_catalog/dbx_schema/bookstore/raw/customers-json/`
```

```sql
create or replace temp view customer_new as
select * from json.`/Volumes/dbx_catalog/dbx_schema/bookstore/raw/customers-json-new/`;

MERGE INTO customers a
USING customer_new b
ON a.customer_id = b.customer_id
WHEN MATCHED AND a.email IS NULL and b.email IS NOT NULL THEN UPDATE SET a.email = b.email
WHEN NOT MATCHED THEN INSERT *;
```

```sql
MERGE INTO books a
USING (
    SELECT *, NULL as _rescued_data from books_new
)  b
ON a.book_id = b.book_id and a.title = b.title
WHEN NOT MATCHED and b.category = 'Computer Science'  THEN INSERT *
```
